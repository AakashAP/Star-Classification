### IMPORT LIBRARY

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plot
```

### IMPORT DATASET

```python
Dataset = pd.read_csv('Dataset/Dataset.csv')
```

DATASET EXPLORATION

```python
print("Dataset : ", Dataset.shape)
print(Dataset.shape[0], "Entries with", Dataset.shape[1], "Columns")

Dataset.head()
```

---
### DATASET TRANSFORMATION

```python
Dataset.drop(columns=['Object_ID', 'Plate', 'Run_ID', 'Rerun_ID', 'CamCol_ID', 'Field_ID', 'Fiber_ID', 'SpecObj_ID', 'MJD'], inplace=True)
```

```python
Dataset = Dataset[
    (Dataset["Alpha"] >= 0) & (Dataset["Alpha"] <= 360) &
    (Dataset["Delta"] >= 0) & (Dataset["Delta"] <= 360)
]
```

```python
print("Modified Dataset : ", Dataset.shape)
print(Dataset.shape[0], "Entries with", Dataset.shape[1], "Columns")

Dataset.head()
```

DATASET SPLITTING

```python
In = Dataset.drop(columns=['Target'])
Out = Dataset.loc[:, 'Target'].values
```

```python
figure = plot.figure(figsize=(10, 8))

sns.heatmap(In.corr(), annot=True)
plot.show()
```

```python
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter


Before = Counter(Out)

Sampler = RandomUnderSampler(sampling_strategy='majority')
In, Out = Sampler.fit_resample(In, Out)

After = Counter(Out)

figure, axes = plot.subplots(1, 2, figsize=(8, 4))
axes[0].bar(Before.keys(), Before.values())
axes[1].bar(After.keys(), After.values())

plot.show()
```

DATASET ENCODING

```python
from sklearn.preprocessing import LabelEncoder


LabelEncoder = LabelEncoder()
Out = LabelEncoder.fit_transform(Out)
```

TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split


TrainIN, TestIN, TrainOUT, TestOUT = train_test_split(In, Out, train_size=0.8, test_size=0.2)

print("Training Set :", len(TrainIN))
print("Testing Set :", len(TestIN))
```

STANDARDIZATION

```python
from sklearn.preprocessing import StandardScaler


Scaler = StandardScaler()

Scaler.fit(TrainIN)
TrainIN = Scaler.transform(TrainIN)

Scaler.fit(TestIN)
TestIN = Scaler.transform(TestIN)
```

---
### MACHINE LEARNING MODEL
MODEL CREATION

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier


AI_Model = {
    "Support Vector Machine"  : SVC(),
    "K-Neighbors Classifier"  : KNeighborsClassifier(),
    "Random Forest Clasifier" : RandomForestClassifier()
}
```

MODEL TRAINING

```python
for Model in AI_Model.values():
    Model.fit(TrainIN, TrainOUT)
```

MODEL EVALUATION

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report


for Name, Model in AI_Model.items():
    Pred = Model.predict(TestIN)

    print(f'For {Name} :')
    print('\tAccuracy Score   :', round(accuracy_score(TestOUT, Pred), 4)*100, "%")
    print('\tPrecision Score  :', round(precision_score(TestOUT, Pred, average='weighted', zero_division=1), 4)*100, "%")
    print('\tRecall Score     :', round(recall_score(TestOUT, Pred, average='weighted', zero_division=1), 4)*100, "%")

    print('\n\tClassification Report :\n', classification_report(TestOUT, Pred))
    print('\n')


figure, axes = plot.subplots(1,3, figsize=(14, 3))


for i, Model in enumerate(AI_Model.values()):
    Pred = Model.predict(TestIN)
    sns.heatmap(confusion_matrix(TestOUT, Pred), annot=True, fmt='d', cmap='Blues', ax=axes[i])

plot.show()
```
